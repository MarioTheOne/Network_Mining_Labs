#### Graph objects can be created in NetworkX in 3 ways:
    
• Adding edges and nodes explicitly.

• Importing data from pre-existing sources such as GML, GraphML formats

• Graph generators—standard algorithms to create network topologies, such as
  binomial_graph() or erdos_renyi_graph()

### Adding node ‘one at a time’ or ‘from a node list’:

```python
G = nx.Graph()
G.add_node('a')
G.add_nodes_from(['b', 'c', 'd'])
G.add_nodes_from('xyz')
H = nx.path_graph(5)
G.add_nodes_from(H)
G.nodes()
```

### Adding edges from one at atime or from an edge list

```python
G = nx.Graph([('a','b'), ('b','c'), ('c','a')])
G.add_edge('a', 'd')
G.add_edges_from([('d', 'c'),('d', 'b')])
G.edges()
```

### Adding node attributes

```python
G = nx.Graph()
G.add_node(1, name='Obrian')
G.add_nodes_from([2], name='Romeo')
G.node[1]['name']
```

### Adding edge attributes

```python
G.add_edge(1, 2, w=3.8)
G.add_edges_from([(3, 4), (4, 5)], w=3.0)
#G.add_edges_from([(1, 2, {'val'}: 2.0)]) Not working
G.add_weighted_edges_from([(6, 7, 3.0)])

# Getting edge data
G.get_edge_data(1, 2)
G[6][7]
```

### Simple graph properties

```python
G = nx.Graph()
N = (['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])
E = ([('A', 'B'), ('A', 'C'), ('B', 'D'), ('B', 'G'), ('B', 'H'), ('B', 'F'), ('C', 'E'),
      ('C', 'G'), ('E', 'F')])
G.add_nodes_from(N)
G.add_edges_from(E)
```

#### Number of nodes

```python
len(G)
G.number_of_nodes()
G.order()
```

#### Number of edges

```python
G.number_of_edges()
G.size()
```

#### Checking membership

```python
G.has_node('A')
G.has_edge('A', 'B')
```

#### Listing nodes and edges

```python
list(G.edges)
list(G.nodes)
```

#### Neighbours

```python
list(G.adj['A'])
list(G.neighbors('A'))
```

#### Removing nodes and edges

```python
G.remove_node('A')
G.remove_nodes_from(['B'])
list(G.nodes)
```

```python
G.remove_edge('C', 'E')
list(G.edges())
```

#### Degree

```python
G.degree('C')
G.degree(['C', 'E'])
G.degree() # All degrees

[b for a,b in G.degree()]
```

#### Neighbors iteration

```python
G = nx.Graph()
G.add_path([0, 1, 2, 3, 4])
[(n, nbrs) for n, nbrs in G.adjacency()]
```
